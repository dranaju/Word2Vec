```python
import re
class sentences():
    def _init_ (self):
        pass
        
    def __iter__(self):
        for line in tweets_data:
            splitted_line = line.split(' ')
            sentence = []
            for word in splitted_line:
                if (len(word) > 0):
                    word = word.lower()
                    word = re.sub(r'[_]','',word)
                    word = re.sub(r'@[A-Za-z0-9]+','',word)
                    word = re.sub('https?://[A-Za-z0-9./]+','',word)
                    word = re.sub(r'[":,.!?]','',word)
                    word = re.sub(r'\n','',word)
                    word = re.sub(r'\[','',word)
                    word = re.sub(r'\]','',word)
                    word = re.sub(u'\-','',word)
                    word = re.sub(u'\u2026','',word)
                    word = re.sub(r'\#[A-Za-z0-9]+','',word)
                    word = re.sub(u'\xe2\x80\xa6','', word)
                    word = re.sub(u'\xf0\x9f\x98\x86','',word)
                    if word != 'rt' and word != ':' and word != '':
                        sentence.append(word)
            yield sentence
```

```python
points = pd.DataFrame(
    [
        (word, coords[0], coords[1])
        for word, coords in [
            (word, all_word_vectors_matrix_2d[model.wv.vocab[word].index])
            for word in model.wv.vocab
        ]
    ],
    columns=["palavra", "x", "y"]
)
```

```python
def  plot_regionplot_reg (x_bounds, y_bounds):
    slice = points[
        (x_bounds[0] <= points.x) &
        (points.x <= x_bounds[1]) & 
        (y_bounds[0] <= points.y) &
        (points.y <= y_bounds[1])
    ]
    
    ax = slice.plot.scatter("x", "y", s=35, figsize=(10, 8))
    for i, point in slice.iterrows():
        try:
            ax.text(point.x + 0.005, point.y + 0.005, point.palavra, fontsize=11)
        except:
            continue
```

```python
plot_regionplot_reg (x_bounds=(0.2,0.3), y_bounds=(-0.4, -0.3))
```
